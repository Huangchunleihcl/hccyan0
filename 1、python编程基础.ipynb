## 【代码解释】

第1~3行导入3个必要的包，并分别取一个简化的别名

第6行是一种特殊用法,它告知Jupyter将Matplotlib图标直接在浏览器中显示

第7行定义一个2*30的矩阵,矩阵的元素是随机产生的,并将其值给变量a

第8行启动一个TensorFlow的Session,并将其赋值给sess对象

第9行调用sess对象的run()方法并将a作为run方法的参数,同时将返回值赋值给out对象

第10行,将2*30的矩阵分为两个1*15的向量,分别赋值给x和y

第11行,以分别x和y做横、纵坐标,利用pyplot绘制散点图第10行显示散点图

```python
#---------------------------------------------测试练习10.3

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
tf.compat.v1.disable_eager_execution()
#---保证sess.run()能够正常运行


%matplotlib inline
#---功能是内嵌绘图，可以省略plt.show()这一步
a = tf.random.normal([2,30])
sess = tf.compat.v1.Session()

out = sess.run(a)
x,y = out
plt.scatter(x,y)

plt.show()
```

# 2、Scipy

```python
from scipy.sparse  import coo_matrix, csr_matrix, csc_matrix
import numpy as np
I = np.array([0, 0, 3, 1, 0])
J = np.array([0, 0, 3, 1, 2])
V = np.array([4, 2, 5, 7, 9])
A = coo_matrix((V, (I, J)), shape=(4, 4))
B = csr_matrix((V, (I, J)), shape=(4, 4))
C = csc_matrix((V, (I, J)), shape=(4, 4))
print(A)
print(B)
print(C)
```

### 1.2 if条件语句 

```python
import random  # 随机模块

player = int(input("请输入（0-剪刀  1-石头 2-布）"))  # 玩家输入
computer = random.randint(0,2)   # 0-2之间随机产生一个数作为电脑的输入

if (player == 0 and computer == 2) or (player == 1 and computer == 0) or (player == 2 and computer == 1):
    print("恭喜玩家你赢了")  # 加上小括号
elif (player == 0 and computer == 0) or (player == 1 and computer == 1) or (player == 2 and computer == 2):
    print("平局")
else:
    print("很可惜，你输了")
```
